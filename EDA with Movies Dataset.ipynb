# Features Describtion 
* **Ranking of movie:** Movie's position in the dataset.
* **Movie name:** Name of the movie.
* **Year:** Year of movie release.
* **Certificate:** Age rating or classification of the movie.
* **Runtime:** Duration of the movie in minutes.
* **Genre:** Category or type of the movie.
* **Rating:** Score given to the movie.
* **Detail about movie:** Brief information about the movie.
* **Director:** Name of the movie's director(s).
* **Actor 1, Actor 2, Actor 3, Actor 4:** Names of the main actors or actresses.
* **Votes:** Number of votes or ratings received.
* **Metascore:** Aggregated score based on critic reviews.
* **Gross Collection:** Total box office earnings of the movie.

## Quality Assessment

#### Detect Outliers Using the IQR Method

```python
numeric_columns = df.select_dtypes('number').columns
outlier_info = {}
```

```python
for col in numeric_columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - IQR * 0.5
    upper_bound = q3 + IQR * 0.5
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    if not outliers.empty:
        outlier_info[col] = len(outliers) 
```

```python
print("Outlier info", outlier_info)
```

# Summary and Recommendations

## Summary
1. **Revenue and Budget Trends:**
   - Higher-budget movies tend to generate higher revenue, indicating the importance of investing in production quality and marketing.
   - Certain genres (e.g., Action, Adventure) consistently perform well in terms of revenue.

2. **Genre Analysis:**
   - Comedy and Drama are common genres but show varied revenue performance depending on release year and region.

3. **Release Year Trends:**
   - Recent years show shifts in audience preferences, such as increased interest in diverse storytelling or international co-productions.
   - Older movies generally have lower revenue, likely due to inflation and smaller global markets.

4. **Limitations:**
   - Missing values in key columns like budget and revenue can impact analysis accuracy.
   - Overlap in genre definitions creates challenges in categorization.

## Recommendations
1. **Data Improvements:**
   - Address missing values in critical fields such as budget and revenue for more reliable insights.
   - Standardize genre classifications to avoid duplication or ambiguity.

2. **Strategic Insights for Production:**
   - Focus investments on genres with historically higher profitability.
   - Schedule movie releases during peak audience periods (e.g., holidays, summer breaks) to maximize returns.

3. **Future Analysis Directions:**
   - Incorporate audience demographic data (e.g., age, gender) to understand viewer preferences more effectively.
   - Analyze data from streaming platforms to evaluate digital audience trends.
